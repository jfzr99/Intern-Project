```python
import pandas as pd

# Read CSV files into DataFrames
df1 = pd.read_csv('C://Users//Ajeng//Documents//Project Intern Xeratic//final project repo//content.csv')
df2 = pd.read_csv('C://Users//Ajeng//Documents//Project Intern Xeratic//final project repo//context.csv')


if df1.shape[0] != df2.shape[0]:
    raise ValueError("The number of rows in the two files does not match.")

# Concatenate the DataFrames side by side
data = pd.concat([df1, df2], axis=1)
display(data)

#data.to_excel("C://Users//Ajeng//Documents//Project Intern Xeratic//final project repo//test.xlsx")
```

```python
data = data.drop(data.columns[0], axis=1)
```

```python
display(data)
```

add column pasal

```python
# Initialize the new column 'pasal' with None values
data['pasal'] = None

# Iterate over the dataframe to fill in the 'pasal' column based on 'context'
current_pasal = None
for index, row in data.iterrows():
    if row['context'] == 'Pasal':
        current_pasal = row['context_number']
    data.at[index, 'pasal'] = current_pasal

# Show the modified DataFrame to confirm the changes
data.head(10)
```

```python
#create multiple values in the new column with multiple conditions
import numpy as np
conditions = [
    (data['pasal'] == '80'),
    (data['pasal'] == '81'),
    (data['pasal'] == '82'),
    (data['pasal'] == '83'),
    (data['pasal'] == '84')
]
values = ['Kesatu', 'Kedua', 'Ketiga', 'Keempat', 'Kelima']

# Apply conditions and assign the result to a new column 'bagian'
data['bagian'] = np.select(conditions, values, default=None)
data.head()
```
